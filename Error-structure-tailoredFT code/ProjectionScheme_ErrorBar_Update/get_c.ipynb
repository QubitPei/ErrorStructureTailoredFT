```python
import pickle
import matplotlib.pyplot as plt

p_dep = 1e-3
phi = 1e-3

k_list = [3, 4, 5, 6]
```

```python

Output_Filename = "MainProjScale251216m2.pkl"
with open(Output_Filename, 'rb') as file:
    data = pickle.load(file)
tr_list =data['Result_Disp_List'][:,0]
c_k_list = [tr_list[i] / (p_dep ** 2 * phi) for i in range(len(k_list))]

Output_Filename = "MainProjScale251219m2.pkl"
with open(Output_Filename, 'rb') as file:
    data = pickle.load(file)
tr_list =data['Result_Disp_List'][:,0]
c_k_list2 = [tr_list[i] / (p_dep ** 2 * phi) for i in range(len(k_list))]
```

```python
plt.plot(k_list, c_k_list, marker='o')
plt.plot(k_list, c_k_list2, marker='D')
plt.xlabel(r'$k$')
plt.ylabel(r'$c(k) = D_{tr}(k)/(p_{dep}^2 * \phi)$')
plt.title('m=2, num_shots={}'.format(data['num_shots']))
```
